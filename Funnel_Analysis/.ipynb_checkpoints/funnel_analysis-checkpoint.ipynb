# A full picture of funnel conversion rate for both desktop and mobile

```python
user['date']=pd.to_datetime(user['date'])
user['user_id'] = user['user_id'].astype('object')
```

```python
user.describe(include='all')
```

```python
home_page['user_id'] = home_page['user_id'].astype('object')
home_page.columns =['user_id', 'home_page']
home_page['home_page'] =1 
home_page.dtypes
```

```python
home_page.describe(include='all')
```

```python
search_page['user_id'] = search_page['user_id'].astype('object')
search_page.columns =['user_id', 'search_page']
search_page['search_page'] =1 
search_page.dtypes
```

```python
search_page.describe(include='all')
```

```python
payment_page['user_id'] = payment_page['user_id'].astype('object')
payment_page.columns =['user_id', 'payment_page']
payment_page['payment_page'] =1 
payment_page.dtypes
```

```python
payment_page.describe(include='all')
```

```python
payment_confirm['user_id'] = payment_confirm['user_id'].astype('object')
payment_confirm.columns =['user_id', 'payment_confirm']
payment_confirm['payment_confirm'] =1 
payment_confirm.dtypes
```

```python
payment_confirm.describe(include='all')
```

```python
# table is clean - no dup records
```

```python
merge_df = user.merge(home_page, how='left', on ='user_id').\
merge(search_page, how='left', on ='user_id').\
merge(payment_page, how='left', on ='user_id').\
merge(payment_confirm, how='left', on ='user_id')
merge_df.head()
```

```python
conversion = merge_df.groupby('device').agg({'user_id':'count','home_page':'sum','search_page':'sum', 'payment_page':'sum', 'payment_confirm':'sum'}).reset_index()
conversion.columns = ['device','traffic','home_page_conversions','search_page_conversions','payment_page_conversions','payment_confirm_conversions']
# conversion['cvr'] = conversion['conversions']/conversion['traffic']
conversion
```

```python
# since traffic and home_page_conversion are the same, drop traffic 
conversion.drop(columns = ['traffic'], inplace=True)
```

```python
conversion.columns
```

```python
conversion['homepage_search'] = conversion['search_page_conversions']/conversion['home_page_conversions']
conversion['search_payment'] = conversion['payment_page_conversions']/conversion['search_page_conversions']
conversion['payment_payconfirm'] = conversion['payment_confirm_conversions']/conversion['payment_page_conversions']
```

```python
conversion
```

```python
merge_df.query('home_page==1 and search_page==1 and payment_page==1 and payment_confirm==1').shape
# check if users need to go thr each step to finally convert
```

```python
 merge_df.shape
```

```python
conversion_dat = merge_df.groupby(['device','date']).agg({'home_page':'sum','search_page':'sum', 'payment_page':'sum', 'payment_confirm':'sum'}).reset_index()
conversion_dat.columns = ['device','date','home_page_conversions','search_page_conversions','payment_page_conversions','payment_confirm_conversions']

conversion_dat
```

```python
conversion.columns
```

```python
conversion['cvr%']=conversion['payment_confirm_conversions']*100/conversion['home_page_conversions']
conversion[['device','cvr%']]
```

```python
merge_df.fillna(0,inplace=True)
merge_df.head()
```

```python
# alternative approach
#  We can safely assume that people who made it to the search page have some intent to buy. 
# However, only 1% of these people convert on desktop and 2% of mobile
# The percentage of bouncers (i.e. they leave after landing on the home page) is actually the same.
merge_df.groupby(['device']).\
apply(lambda x: pd.Series({'to_search':x['search_page'].mean()/x['home_page'].mean()\
                    ,'to_payment':x['payment_page'].mean()/x['search_page'].mean()
                    ,'to_confirm':x['payment_confirm'].mean()/x['payment_page'].mean()
                          })).reset_index()
```

```python
conversion_dat.columns
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(conversion_dat, x='date',y='home_page_conversions',hue='device')
plt.title('home page by device')

plt.figure(figsize=(10, 5))
sns.lineplot(conversion_dat, x='date',y='search_page_conversions',hue='device')
plt.title('search page by device')


plt.figure(figsize=(10, 5))
sns.lineplot(conversion_dat, x='date',y='payment_page_conversions',hue='device')
plt.title('payment page by device')

plt.figure(figsize=(10, 5))
sns.lineplot(conversion_dat, x='date',y='payment_confirm_conversions',hue='device')
plt.title('confirmation page by device')
```

## look at by sex 

```python
merge_df.head()
```

```python
 merge_df.groupby(['sex'])['payment_confirm'].agg({'mean','count'})
```

```python
test = stats.ttest_ind(merge_df.query("sex == 'Female'")['payment_confirm'],merge_df.query("sex == 'Male'")['payment_confirm'], equal_var = False)
```

```python
test.pvalue # sex doesn’t affect conversion rate in a statistically significant way
```

```python
merge_df.groupby(['sex','device'])['payment_confirm'].agg({'mean','count'})
```

```python
conversion_sex = merge_df.groupby(['date','sex']).apply(lambda x:\
    pd.Series({'to_search':x['search_page'].mean()/x['home_page'].mean()\
              ,'to_payment':x['payment_page'].mean()/x['search_page'].mean()\
               ,'to_confirm':x['payment_confirm'].mean()/x['payment_page'].mean()
              })).reset_index()
conversion_sex.head()
```

```python
conversion_sex.columns
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(conversion_sex, x='date',y='to_search',hue='sex')
plt.title('home page to search by sex')

plt.figure(figsize=(10, 5))
sns.lineplot(conversion_sex, x='date',y='to_payment',hue='sex')
plt.title('search to payment page by sex')


plt.figure(figsize=(10, 5))
sns.lineplot(conversion_sex, x='date',y='to_search',hue='sex')
plt.title('payment page by sex')

plt.figure(figsize=(10, 5))
sns.lineplot(conversion_sex, x='date',y='to_confirm',hue='sex')
plt.title('payment to confirm page by sex')
```
